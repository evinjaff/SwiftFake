```python
import json
import ollama



json.dump(eval_results, open("eval_results_benign.json", "w"))
json.dump(eval_results_aid, open("eval_results_aid_benign.json", "w"))




for i_eval_results in [eval_results, eval_results_aid]:
    hard_block_calls = 0
    hard_allow_calls = 0
    soft_block_calls = 0
    soft_allow_calls = 0
    ambiguous_calls = 0

    for eval in i_eval_results:


        if "call: block" in eval["closing"]:
            hard_block_calls += 1
        elif "call: allow" in eval["closing"]:
            hard_allow_calls += 1
        
        else:
            # run ollama with llama3.2:3b to summarize it
            result = ollama.chat(model="llama3.2:3b", messages= [ {
                "role": "user",
                "content": "The following is a report about a call and whether it should be blocked or allowed. \
                Please summarize the call and provide a recommendation for whether the call should be blocked or allowed. \
                ONLY print call: block or call: allow. {}".format(eval["transcript"])
            } ] )

            output = result["message"]["content"]

            if "call: block" in output:
                soft_block_calls += 1
            elif "call: allow" in output:
                soft_allow_calls += 1
            else:
                ambiguous_calls += 1



    print(hard_block_calls, hard_allow_calls, soft_block_calls, soft_allow_calls, ambiguous_calls)
        
    




```
