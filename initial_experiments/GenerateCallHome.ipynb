```python

import numpy as np
import soundfile as sf

def construct_audio(entry, filename):

    # construct the audio from array
    bytes = entry["audio"]["array"]

    sample_rate = entry["audio"]["sampling_rate"]

    timestamps = entry["timestamps_start"]
    speakers = entry["speakers"]

    # get sequence of each speaker
    speaker_a = []
    speaker_b = []

    print(timestamps)
    print(type(timestamps))
    print(speakers)
    print(type(speakers))

    
    def get_speaker_sequence(speakers, timestamps, speaker):
        start = None
        sequence = []
        for i in range(len(speakers)):
            if speakers[i] == speaker:
                if start is None:
                    start = timestamps[i]
            else:
                if start is not None:
                    sequence.append((start, timestamps[i]))
                    start = None
        if start is not None:
            sequence.append((start, timestamps[-1]))
        return sequence

    speaker_a = get_speaker_sequence(speakers, timestamps, "A")
    speaker_b = get_speaker_sequence(speakers, timestamps, "B")


    print(len(speaker_a), len(speaker_b))
    print(speaker_a, speaker_b)

    speaker_a_first = speaker_a[0]
    speaker_b_first = speaker_b[0]

    # get the audio for the first speaker A sequence
    start_a = int(speaker_a_first[0] * sample_rate)
    end_a = int(speaker_a_first[1] * sample_rate * 1.2)
    start_b = int(speaker_b_first[0] * sample_rate)
    end_b = int(speaker_b_first[1] * sample_rate * 1.2)

    audio_a = bytes[start_a:end_a]
    audio_b = bytes[start_b:end_b]

    print(len(audio_a), len(audio_b))

    # write the audio to a file
    sf.write("audio_a.wav", audio_a, sample_rate)
    sf.write("audio_b.wav", audio_b, sample_rate)

    # save the longer audio to a file
    if len(audio_a) > len(audio_b):
        sf.write("{}.wav".format(filename), audio_a, sample_rate)
    else:
        sf.write("{}.wav".format(filename), audio_b, sample_rate)


for i in range(len(ds["data"])):
    entry = ds["data"][i]
    construct_audio(entry, "CallHomeOpenings/audio_{}".format(i))

# entry = ds["data"][0]
# construct_audio(entry, "CallHomeOpenings/audio_0")

```
